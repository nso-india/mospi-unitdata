# mospi-unitdata Demo

Demonstrates the programmatic API added in v0.2.0.

```python
API_KEY = "YOUR_API_KEY"  # Replace with your key from https://microdata.gov.in
```

## List datasets (first page)

```python
from MospiUnitdata import list_datasets

datasets = list_datasets(API_KEY, page=1)
for d in datasets:
    print(f"{d['id']:>4} | {d['idno']:<45} | {d['title']}")
```

## Search datasets by keyword

```python
results = list_datasets(API_KEY, query="labour force")
print(f"Found {len(results)} datasets:\n")
for d in results:
    print(f"  {d['id']} ({d['idno']}): {d['title']}")
```

## List files in a dataset

```python
from MospiUnitdata import list_files

files = list_files("DDI-IND-NSO-ASI-2020-21", API_KEY)
print(f"{len(files)} files available:\n")
for f in files:
    print(f"  {f['name']:<50} {f.get('size', '?')}")
```

## Download a single file

```python
from MospiUnitdata import download_file

path = download_file("DDI-IND-NSO-ASI-2020-21", "ASI_DATA_2020_21_CSV.zip", "./data", API_KEY)
```

## Download all files in a dataset

```python
from MospiUnitdata import download_dataset

downloaded = download_dataset("DDI-IND-NSO-ASI-2020-21", "./data", API_KEY)
print(f"\nDownloaded {len(downloaded)} files")
```
